```python
# Contact Book:-

phonebook = {}  # Format: { "Name": {"phone": "...", "email": "...", "address": "..."} }

def list_contacts():#Print all contacts in a table.
    if not phonebook:
        print("\n Your contact list is empty!\n")
        return
    print("\n======= Your Contacts =======")
    print(f"{'Name':<18}{'Phone':<15}{'Email':<28}{'Address'}")
    print("-" * 80)
    for person, details in phonebook.items():
        print(f"{person:<18}{details['phone']:<15}{details['email']:<28}{details['address']}")
    print("-" * 80 + "\n")

def create_contact():#Add a new contact to the phonebook.
    name = input("Enter full name: ").strip()
    if name in phonebook:
        print("⚠ Contact already exists. Try updating it instead.\n")
        return
    phone = input("Enter phone number: ").strip()
    email = input("Enter email: ").strip()
    address = input("Enter address: ").strip()
    phonebook[name] = {"phone": phone, "email": email, "address": address}
    print(f"Contact '{name}' saved.\n")

def search_contacts():#Find a contact by name or phone number.
    term = input("Search by name or phone: ").strip()
    matches = [
        (name, info) for name, info in phonebook.items()
        if term.lower() in name.lower() or term == info['phone']
    ]
    if matches:
        print("\n Search Results:")
        for name, details in matches:
            print(f"Name   : {name}")
            print(f"Phone  : {details['phone']}")
            print(f"Email  : {details['email']}")
            print(f"Address: {details['address']}\n")
    else:
        print(" No contacts found for your search.\n")

def update_contact():#Edit an existing contact's details.
    name = input("Enter the name of the contact to edit: ").strip()
    if name not in phonebook:
        print(" Contact not found.\n")
        return
    print("Press Enter to keep the existing value.")
    phone = input(f"New phone ({phonebook[name]['phone']}): ").strip() or phonebook[name]['phone']
    email = input(f"New email ({phonebook[name]['email']}): ").strip() or phonebook[name]['email']
    address = input(f"New address ({phonebook[name]['address']}): ").strip() or phonebook[name]['address']
    phonebook[name] = {"phone": phone, "email": email, "address": address}
    print(f" Contact '{name}' updated.\n")

def remove_contact():#Delete a contact permanently
    name = input("Enter the contact name to delete: ").strip()
    if name in phonebook:
        confirm = input(f"Are you sure you want to delete '{name}'? (y/n): ").strip().lower()
        if confirm == 'y':
            del phonebook[name]
            print("Contact removed.\n")
    else:
        print("Contact not found.\n")
# Program Menu
while True:
    print("""
 CONTACT BOOK MENU
1. Add Contact
2. Search Contact
3. View All Contacts
4. Edit Contact
5. Delete Contact
6. Exit
""")
    option = input("Choose an option (1-6): ").strip()

    if option == "1":
        create_contact()
    elif option == "2":
        search_contacts()
    elif option == "3":
        list_contacts()
    elif option == "4":
        update_contact()
    elif option == "5":
        remove_contact()
    elif option == "6":
        print(" Goodbye! Your contacts are safe (for now).")
        break
    else:
        print("⚠ Invalid choice. Please pick 1–6.\n")
```
